```
df_final_customer = pd.read_csv('./all_customer.csv', encoding='utf-8-sig')
df_final_card = pd.read_excel('./merged/merged_card.xlsx')
```

```
def filter_by_previous(df_customer, df_card, ID):
    # 기준금액 가져오기 (ID에 해당하는 고객)
    base_amount = df_customer[df_customer['ID'] == ID]['이용금액_신판_B0M'].values[0] * 1000
    # 조건에 맞는 행만 남기기
    filtered_df = df_card[df_card['직전 1개월 합계 실적 금액(만원)'] * 10000 <= base_amount]
    return filtered_df

customer_dict = {}
for IDs in df_final_customer['ID']:
        customer_dict[IDs] = filter_by_previous(df_final_customer, df_final_card.copy(), IDs)

```
